```python
import pandas as pd
import os
```

```python
FOOD_PATH = "datasets/food"
```

```python
def load_food_prices_data(food_path=FOOD_PATH):
    csv_path = os.path.join(food_path,'wfp_market_food_prices.csv')
    return pd.read_csv(csv_path, encoding='latin-1')
```

```python
food = load_food_prices_data()
```

```python
food.columns = ['country_id','country_name', 'locality_id','locality_name','market_id','market_name', 'commodity_purchase_id','commodity_purchased','currency_id','currency_name','market_type_id','market_type','measurement_id','measurement_unit','month_recorded','year_recorded','price_paid','commodity_source']
```
